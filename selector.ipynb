```python
import pandas as pd
from sklearn.preprocessing import StandardScaler, MinMaxScaler
from sklearn.feature_selection import SelectFromModel, SelectKBest, chi2
from sklearn.linear_model import LinearRegression
from sklearn.feature_selection import SequentialFeatureSelector
from sklearn.ensemble import RandomForestRegressor
df = pd.read_csv('placement.csv')
df1=df
df2=df
```

```python
scaler = StandardScaler()
df[df.columns] = scaler.fit_transform(df[df.columns])
df
```

```python
scaler = MinMaxScaler()
df1[df1.columns] = scaler.fit_transform(df1[df1.columns])
df1
```

```python
selector = SelectFromModel(estimator=LinearRegression(), max_features=1)
selector.fit_transform(df[df.columns[:-1]], df['cgpa'])
```

```python
from sklearn.preprocessing import LabelEncoder
df2 = pd.read_csv('IRIS.csv')
le=LabelEncoder()
le.fit(df2['species'])
df2=df2.dropna()
sfs = SequentialFeatureSelector(RandomForestRegressor(), n_features_to_select=2)
ad=sfs.fit_transform(df2[df2.columns[:-2]], le.transform(df2['species']))
ad
```

```python

selector = SelectKBest(chi2, k=2)
selector.fit_transform(df2[df2.columns[:-1]], df2['species'])
```
