```python
import matplotlib.pyplot as plt
import pandas as pd
data=pd.read_csv('marks.csv')
```

```python
plt.bar(data['Subject'],data['Marks'])
plt.xlabel('Subject')
plt.ylabel('Marks')
plt.title('Marks of Student')
plt.show()
```

```python
plt.pie(data['Marks'],labels=data['Subject'])
plt.title('Marks of Student')
plt.show()
```

```python
plt.hist(data['Marks'])
plt.xlabel('Marks')
plt.ylabel('Frequency')
plt.title('Marks of Student')
```
