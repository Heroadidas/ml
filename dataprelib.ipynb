```python
import numpy as np
import pandas as pd
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import LabelEncoder, OneHotEncoder

df=pd.read_csv('marks2.csv')
df
```

```python
df=df.dropna()
df
```

```python
new_row = pd.DataFrame([{'Subject':'TOC','Marks':np.nan}])
df = pd.concat([df,new_row])
imputer=SimpleImputer(missing_values=np.nan,strategy='mean')
df['Marks']=imputer.fit_transform(df['Marks'].values.reshape(-1,1))
df
```

```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder

df = pd.DataFrame({
    'Color': ['Red', 'Blue', 'Green', 'Red', 'Blue']
})
dff=df
onehotencoder = OneHotEncoder()
X = onehotencoder.fit_transform(df.Color.values.reshape(-1,1)).toarray()
dfOneHot = pd.DataFrame(X, columns = ["Color_"+str(int(i)) for i in range(X.shape[1])]) 
df = pd.concat([df, dfOneHot], axis=1)
df = df.drop(['Color'], axis=1)
df
```

```python
le=LabelEncoder()
le.fit(dff['Color'])
list(le.classes_)
dff['Color']=le.transform(dff['Color'])
dff
```
